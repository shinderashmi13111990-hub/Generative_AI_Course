# Sets

#### A set in Python is an unordered, mutable collection of unique, hashable elements.

### Core Characteristics
##### Unique Elements: Duplicate values are automatically ignored.
##### Unordered: Items do not maintain any specific sequence or index.
##### Unindexed: Elements cannot be accessed via bracket notation (my_set[0] raises a TypeError).
##### Mixed Types: Sets can contain distinct immutable types like strings, integers, and tuples.
##### Mutable Structure: You can add or remove items, but the individual elements themselves must be immutable (e.g., you cannot place a list inside a set)

```python
sets={1,2,3,4}
sets
```

```python
birds={"eagle","hawk"}
birds
```

```python
# Adding an element to the set
birds.add("sparrow")        
birds
```

```python
# Removing an element from the set
birds.remove("pigeon")  # This will raise a KeyError if "pigeon" is not in the set
birds
```

```python
# Using discard to remove an element from the set   
birds.discard("pigeon")  # This will not raise an error if "pigeon" is not in the set
birds
```

```python
random_bird=birds.pop()  # This will remove and return an arbitrary element from the set
print("Randomly removed bird:", random_bird)
```

```python
birds.clear()  # This will remove all elements from the set
birds
```

```python
print(len(birds))  # This will print the number of elements in the set
```

```python
# Python sets excel at mathematical comparisons. 
# You can perform these operations using either methods or shorthand operators

set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

print(set_a | set_b)  # Union of set_a and set_b
print(set_a & set_b)  # Intersection of set_a and set_b
print(set_a - set_b)  # Difference of set_a and set_b
print(set_a ^ set_b)  # Symmetric difference of set_a and set_b
```
