# Dictionary

### Dictionary is a built-in data structure that stores data in key-value pairs, functioning like a hash map. 
### Dictionaries are mutable (changeable), guarantee insertion order , and require unique keys.

### Core Characteristics
##### Keys: Must be unique and immutable data types, such as strings, numbers, or tuples.
##### Values: Can be of any data type, can repeat, and can be modified in place.
##### Performance: Optimized for fast data retrieval with O(1) average time complexity for lookup

```python
dictionary={"Name":"Amit","Phone":"1234","email":"abc@gmail.com","Qualification":"BTech"}

dictionary
```

```python
print(dictionary["Name"])
print(dictionary["Phone"])
print(dictionary["email"])
print(dictionary["Qualification"])
```

```python
dictionary["Name"]="Rashmi"
dictionary
```

```python
dict={"Name":"Amit","Name":"Prasad","Name":"Prakash"}
dict
```

##### Dictionary will accept duplicate keys but return only the last key-vaule pair. Thats why The keys needs to be unique

## Nested Dictionary
#### Nested Dictionary will be used for keeping a dictionary inside another dictionary

```python
Nested_dictionary={"Student1":{"Name":"Amit","Marks":84,"Grade":"A"},
                   "Student2":{"Name":"Ganesh","Marks":96,"Grade":"A+"},
                   "Student3":{"Name":"Rahul","Marks":74,"Grade":"B"}}

Nested_dictionary
```

```python
# Iterating over Nested Dictionary

for outer_key,outer_value in Nested_dictionary.items():
    print(f"Outer key is : {outer_key} and outer_values : {outer_value}")
    for key1,value1 in outer_value.items():
        print(f"Inner key is : {key1} and Inner value is : {value1}")
```

## Dictionary Comprehension
### Dictionary comprehension helps us with using looping statement along with Conditional statement inside our Dictionary

```python
comprehension_dict={x:x**2 for x in range(1,11)}

comprehension_dict
```

```python
comprehension_dict={x:x**2 for x in range(11)}

comprehension_dict
```

```python
comprehension_dict={x:x**2 for x in range(1,11) if x%2==0}
comprehension_dict
```

## Merged Dictionary

```python
# Merging Method 1
dictionary_1={'a':1,'b':2,'c':3}
dictionary_2={'d':4,'e':5,'c':6}

Merged_dictionary1={**dictionary_1,**dictionary_2}
Merged_dictionary1
```

```python
# Merging Method 2
Merged_dictionary2= dictionary_1 | dictionary_2
Merged_dictionary2
```

```python
# Merging Method 3
dictionary_1.update(dictionary_2)
dictionary_1
```
