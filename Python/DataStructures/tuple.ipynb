# Tuples

#### Tuples are the second type of Data Structure which can store the data in a ordered way just like list but tuples are immutable 

```python
tuples_values=(1,2,3)
tuples_values
```

### Methods in tuples
##### count() -- Returns the number of times a specified value occurs in the tuple
##### index() -- Searches the tuple for a specified value and returns the index position of its first occurrence

```python
tuples_values.count(1)
```

```python
tuples_values.index(3)
```

### Slicing

```python
tuples_values[1:4]
```

```python
tuples_values[:4]
```

```python
tuples_values[::-1]
```

### Packing of Tuples

```python
a=1,2,3,4,5,"Amit",6,7

print(type(a))

a
```

### Unpacking of Tuples

```python
a,b,c=1,45,"Amit"

print(a)
print(b)
print(c)
```

## Tuples in Nested List

```python
Tuple_list=[(1,2,3),(4,5,6),(7,8,9)]

print(Tuple_list[0])
print(Tuple_list[1])
print(Tuple_list[2])
```

## Nested Tuples

```python
Nested_tuple=((1,2,3),(4,5,6),(7,8,9))

print(Tuple_list[0])
print(Tuple_list[1])
print(Tuple_list[2])
```

```python
# Iterating over a Nested Tuple

for i in Nested_tuple:
    for j in i:
        print(j, end=" ")
    print(" ")
```

## Operation in Tuples

```python
num_1=40
num_2=34
print(num_1+num_2)
```

```python
## "+" in tuple will concatenate the tuples

tuple_1=(1,2,3)
tuple_2=(4,5,6)
print(tuple_1+tuple_2)
```
