#### Temperature conversion Function

```python
## To create a function that converts temperature from Celcius to Fahrenheit or Celcius to Fahrenheit, we can use the following formula:
## F = (C * 9/5) + 32
## C = (F - 32) * 5/9

def convert_temperature(value, unit):
    if unit == "C":
        fahrenheit = round((value * 9/5) + 32, 2)
        print(f"{value}°C is equal to {fahrenheit}°F")
    elif unit == "F":
        celsius = round((value - 32) * 5/9, 2)
        print(f"{value}°F is equal to {celsius}°C")
    else:
        raise ValueError("Unit must be 'C' for Celsius or 'F' for Fahrenheit.") 

```

```python
convert_temperature(1, "C") 
```

```python
convert_temperature(1, "F") 
```

#### Password Strength Checker Function

```python
# Function to check Password Strength
'''
Strong Password Criteria:
1. At least 8 characters long
2. Contains at least one uppercase letter
3. Contains at least one lowercase letter
4. Contains at least one digit
5. Contains at least one special character (!@#$%^&*()-_+=)

Weak Password Criteria:
Any password that does not meet the above criteria is considered weak.
'''

def check_password_strength(password):
    if len(password) < 8:
        return "Weak Password: Must be at least 8 characters long."
    if not any(char.isupper() for char in password):
        return "Weak Password: Must contain at least one uppercase letter."
    if not any(char.islower() for char in password):
        return "Weak Password: Must contain at least one lowercase letter."
    if not any(char.isdigit() for char in password):
        return "Weak Password: Must contain at least one digit."
    if not any(char in "!@#$%^&*()-_+=" for char in password):
        return "Weak Password: Must contain at least one special character (!@#$%^&*()-_+=)."
    return "Strong Password: Meets all criteria."
```

```python
check_password_strength("Hello")
```

```python
check_password_strength("HelloRashmi")
```

```python
check_password_strength("HelloRashmi123")
```

```python
check_password_strength("HelloRashmi123!")
```

#### Function to check Palindrome String

```python
# Function to check if a string is a palindrome
def is_palindrome(string):
    # Remove spaces and convert to lowercase for uniformity
    s = string.lower().replace(" ", "")
    if s == s[::-1]:
        print("The string is a palindrome.")
    else:
        print("The string is not a palindrome.")
```

```python
is_palindrome("ABa")
```

```python
is_palindrome("AABaa")
```

```python
is_palindrome("hello")
```

```python
is_palindrome("A man a plan a canal Panama")
```

#### Function to calculate total cost of items in a shopping cart

```python
# Function to calculate total price of items in a shopping cart

cart = [
    {"name": "Apple", "price": 5, "quantity": 20},
    {"name": "Banana", "price": 3, "quantity": 10},
    {"name": "Orange", "price": 7, "quantity": 10}
]
def calculate_total_price(cart):
    total_price = 0
    for item in cart:
        total_price += item['price'] * item['quantity']
    print(f"The total price of items in the cart is: ${total_price}")

calculate_total_price(cart)
```
