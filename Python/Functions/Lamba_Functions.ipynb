# Lamda Functions

#### A lambda function in Python is a small, anonymous function that can have any number of arguments but can only contain a single expression.

#### Syntax -  lambda arguments: expression

```python
# Function to add five to a number
def add_five(x):
    return x + 5

# Lambda function equivalent
add_five_lambda = lambda x: x + 5

print(add_five(10))        # Output: 15
print(add_five_lambda(10)) # Output: 15
```

### Lambda Function can have any number of arguments but only one expression or logic is allowed
### They are used for Short Operations or as Arguments to Higher Order Functions

### It is function without name and they dont require the return statement

```python
# Lambda function to check if a number is even
is_even = lambda x: x % 2 == 0

print(is_even(4))  
print(is_even(7))
```

```python
# Lambda function to calculate the square of a number
square = lambda x: x ** 2

print(square(5))
print(square(9))
```

```python
# Lambda function to calculate sum of two numbers
sum_lambda = lambda a, b: a + b

print(sum_lambda(3, 7))
print(sum_lambda(10, 15))
```
