# Map Functions

### Map function applies a given function operation to all items in an iterable or a list and returns map object (an iterator)

### Map function is particularly useful for transforming the data or applying an operation on ALL items of the iterable

### map() is going to help us out with avoiding unnecessary lines of code

### Syntax for map() - list(map(function,iterable))

```python
# Squaring all the numbers in a list
def square(list_items):
    empty_list = []
    for items in list_items:
        squared_item = items ** 2
        empty_list.append(squared_item)
    print(empty_list)
    
square([1, 2, 3, 4, 5])
    
```

```python
# Squaring all the numbers in a list using map function with lambda function
squared_list = list(map(lambda x: x ** 2, [1, 2, 3, 4, 5]))
print(squared_list)
```

```python
# Squaring all the numbers in a list using map function without lambda function
def squared_number(num):
    return num ** 2

print(list(map(squared_number, [1, 2, 3, 4, 5])))
```

```python
# Map function with Multiple Iterables
# Adding two lists element-wise using map function with lambda function 
list1 = [1, 2, 3]
list2 = [4, 5, 6]   
added_list = list(map(lambda x, y: x + y, list1, list2))
print(added_list)
```

```python
# Map function to convert a list of temperatures from Celsius to Fahrenheit
celsius = [0, 10, 20, 30, 40]   
fahrenheit = list(map(lambda x: (x * 9/5) + 32, celsius))
print(fahrenheit)   
```

```python
# Map function to convert all the strings in a list to uppercase
strings = ['hello', 'world', 'python', 'map']
uppercase_strings = list(map(lambda s: s.upper(), strings))
print(uppercase_strings)
```

```python
# Map function to convert all the strings in a list to integers
string_numbers = ['1', '2', '3', '4', '5']
integer_numbers = list(map(lambda s: int(s), string_numbers))
print(integer_numbers)
```

```python
# Retrieve the values inside a dictionary using map function
employees = [
    {'name': 'Alice', 'age': 30, 'department': 'HR'},
    {'name': 'Bob', 'age': 25, 'department': 'IT'},
    {'name': 'Charlie', 'age': 35, 'department': 'Finance'}]

def get_employee_name(employee):
    return employee['name']

def get_employee_age(employee):
    return employee['age']

def get_employee_department(employee):
    return employee['department']

employee_names = list(map(get_employee_name, employees))
print(employee_names)

employee_ages = list(map(get_employee_age, employees))
print(employee_ages)

employee_departments = list(map(get_employee_department, employees))
print(employee_departments)
    
    
```
