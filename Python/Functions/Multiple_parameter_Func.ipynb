# Functions with Multiple Parameters

### Positional arguments:
#### They are passed to a function based purely on their order. Syntax : func(val1, val2)

```python
# Function for addition of numbers
def addition(num1, num2):
    print(f"The sum of {num1} and {num2} is: {num1 + num2}")
```

```python
addition(5, 10)
```

```python
addition(5, 10, 15)
```

## Function with Variable length Arguments


```python
## Function to add multiple numbers
def addition(*num):
    add = 0
    for i in num:
        add += i
    print(f"The sum of {num} is: {add}")
    
addition(5, 10, 15, 233, 23, 45, 67)
```

### Keyword Arguments
#### keyword arguments are passed by explicitly naming the parameter and its value using an equals sign
#### func(param1=val1, param2=val2)

```python
def data(**user_data):
    for key, value in user_data.items():
        print(f"{key}: {value}")
        
data(name="Rashmi", age=30, city="Bangalore", profession="Software Engineer")
```

## When you want to pass both positional and keyword arguments

```python
def variable(*positional_args, **keyword_args):
    for arg in positional_args:
        print("Positional arguments are:", arg) 
    for key, value in keyword_args.items():
        print(f"Key is {key} and value is {value}")
```

```python
variable(1, 2, 3, 23, 34, name="Rashmi", age=30, city="Bangalore", profession="Software Engineer", 100)
```

```python
variable(1, 2, 3, 23, 34, name="Rashmi", age=30, city="Bangalore", profession="Software Engineer")
```

## Function with Default parameter

```python
# Function without default parameters
def greet(name, greeting):
    print(f"{greeting}, {name}!")
```

```python
greet("Rashmi", "Hello")
```

```python
greet("Rashmi")
```

```python
# Function with default parameters
def default_greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")
```

```python
default_greet("Rashmi", "Hi")
```

```python
default_greet("Rashmi")
```
