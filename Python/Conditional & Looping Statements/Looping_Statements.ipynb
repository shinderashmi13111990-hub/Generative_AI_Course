# Looping Statements

# 1. For Loop

#### Looping Statements can be used for iterating over a series of numbers or through iterables or data structures 

### Range()
##### The range() function in Python generates an immutable sequence of numbers


```python
# 1. Single Argument: range(stop)

for i in range(5):
    print(i, end=" ")
```

```python
# 2. Two Arguments: range(start, stop)

for i in range(2, 7):
    print(i, end=" ")
```

```python
# 3. Three Arguments: range(start, stop, step)

for i in range(0, 10, 2):
    print(i, end=" ")
```

```python
# Generating a decreasing sequence using range() with a negative step

for i in range(10, 0, -1):
    print(i, end=" ")
```

### In Looping Statements, we can iterate through iterables like String as well

```python
Name = "Rashmi Bhilare"
for char in Name:
    print(char, end=" ")
```

# 2. While Loop

#### A while loop in Python repeatedly executes a block of code as long as a specified condition remains True

```python
i=0
while i < 5:
    print(i, end=" ")
    i=i+1  # you can also use i+=1 as a shorthand for i=i+1
```

```python
# Print Even numbers from 0 to 20 using while loop
i = 0
while i <= 20:
    print(i, end=" ")
    i += 2  # Increment by 2 to get the next even number
```
