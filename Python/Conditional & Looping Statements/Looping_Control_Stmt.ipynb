# break
### - Terminates the loop entirely
### - Jumps completely out of the loop

```python
for number in range(1, 6):
    if number == 3:
        break  # Completely exit the loop here
    print(number)

print("Loop finished!")
```

# continue 
### - Skips the rest of the current iteration
### - Jumps straight to the next iteration

```python
for number in range(1, 6):
    if number == 3:
        continue  # Skip the rest of this iteration
    print(number)

print("Loop finished!")
```

# pass
### - Does absolutely nothing
### - Serves as a syntax placeholder

```python
for number in range(1, 6):
    if number == 3:
        pass  # Placeholder: do nothing for now, but valid syntax
    print(number)
print("Loop finished!")
```
