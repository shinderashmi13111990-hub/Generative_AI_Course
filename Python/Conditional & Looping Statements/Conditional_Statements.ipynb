# Conditional Statements

#### Conditional statements in Python execute different blocks of code based on whether a specific condition evaluates to True or False. 
#### They control the execution flow of your program using the keywords if, elif (else if), and else
#### Python uses indentation (typically 4 spaces) instead of curly braces {} to define code blocks

```python
x = 10

if x > 5:
    print("x is greater than 5")  # Indented block runs if True
else:
    print("x is 5 or less")        # Indented block runs if False
```

```python
# Simple Calculator using if-elif-else

a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
operation=input("Enter operation (+, -, *, /): ")

if operation == "+":
    print("The addition of two numbers is:", a + b)
elif operation == "-":
    print("The subtraction of two numbers is:", a - b)
elif operation == "*":
    print("The multiplication of two numbers is:", a * b)
elif operation == "/":
    print("The division of two numbers is:", a / b)
else:
    print("Invalid operation")
```

# Nested Conditional Statement

```python
# Check if a year is a leap year

# Get integer input from the user
year = int(input("Enter a year: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, "is a leap year")
        else:
            print(year, "is not a leap year")
    else:
        print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```
