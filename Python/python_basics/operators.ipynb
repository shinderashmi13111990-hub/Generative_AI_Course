# Arithmetic Operators 

```python
a=13
b=5

# + Adds two numbers
print("a+b =", a+b)
# - Subtracts second number from the first
print("a-b =", a-b)
# * Multiplies two numbers
print("a*b =", a*b)
# / Divides first number by the second
print("a/b =", a/b)
# % Modulus operator returns the remainder of the division
print("a%b =", a%b)
# ** Exponentiation operator raises the first number to the power of the second
print("a**b =", a**b)
# // Floor division operator divides the first number by the second and rounds down to the nearest whole number
print("a//b =", a//b)
```

# Comparison Operators

```python
a=23
b=15

# == Equality operator checks if the values of two operands are equal
print("a==b =", a==b)
# != Not equal operator checks if the values of two operands are not equal
print("a!=b =", a!=b)
# > Greater than operator checks if the value of the left operand is greater than the right operand
print("a>b =", a>b)
# < Less than operator checks if the value of the left operand is less than the right operand   
print("a<b =", a<b)
# >= Greater than or equal to operator checks if the value of the left operand is greater than or equal to the right operand
print("a>=b =", a>=b)
# <= Less than or equal to operator checks if the value of the left operand is less than or equal to the right operand
print("a<=b =", a<=b)
```

# Logical Operators

```python
a=23
b=15
c=10

print("a==b and b==c:", a==b and b==c)  # This will return False because a is not equal to c
print("a==b or b==c:", a==b or b==c)   # This will return False because neither a is equal to b nor b is equal to c
print("not a==b:", not a==b)       # This will return True because a is not equal to b
```
