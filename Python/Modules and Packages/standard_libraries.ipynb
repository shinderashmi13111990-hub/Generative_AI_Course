# Standard Libraries in Python

### Standard Library is a vast collection of Modules and packages that come bundled with Python and it provides us with vast range of functionalities

```python
## Math Library -> It is having the functionalities for peforming mathematical calculations

from math import *

num= int(input("Enter the number: "))
print(f"Square root of the number : {sqrt(num)}")
print(f"Factorial of the number : {factorial(num)}")
print(f"Power of the number : {pow(num,2)}")
print(f"Ceiling of the number : {ceil(num)}")
print(f"Floor of the number : {floor(num)}")
print(f"Absolute value of the number : {fabs(num)}")
print(f"Logarithm of the number : {log(num)}")
print(f"Exponential of the number : {exp(num)}")
print(f"Trigonometric sine of the number : {sin(num)}")
```

```python
# Random Library -> It is having the functionalities for generating random numbers
import random

print(random.randint(0,100))
print(random.choice([1,2,3,4,5]))
print(random.choice(['Apple','Banana','Mango','Grapes']))
```

```python
# csv library -> It is having functionalities to work with csv
import csv

with open('data.csv','w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['Name','Age','Qualification'])
    writer.writerow(['Rashmi','35','BTech'])
```

```python
with open('data.csv','r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
# Datetime Library --> It helps with functionalities with respect to time and date

from datetime import *

current_time=datetime.now()
print("Current date and time:",current_time)

yesterday_time=current_time-timedelta(days=1)
print("Yesterday date and time:",yesterday_time)
```

```python
# time --> Time Library provides current time and also it can help us to pause the interpreter for seconds
import time

print(time.time())
time.sleep(5)
print(time.time())
```

```python
# json -> used for parsing, encoding, and manipulating JSON (JavaScript Object Notation) data
import json

json_string = '{"name": "Alice", "is_admin": true, "hobbies": ["coding", "chess"]}'

user_data = json.loads(json_string)

print(user_data["name"])      # Outputs: Alice
print(type(user_data))
```

```python
# re -> Module for Regular Expression. Regular Expression is 
# used to retrieve info based on a specific pattern

import re
text="Hello, My name is Rashmi and my password is Rashmi12345"
pattern=r'\d+'
match=re.search(pattern,text)
print(match.group())
```
