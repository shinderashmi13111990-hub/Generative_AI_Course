# File Handling

### NLP -> Text Data (Sequence Data) --> Text Format File
### Machine Learning Models -> Cat Dog Image Classifier -> png or jpeg -> Binary Format file
### Deep Learning -> object Classifier -> png or jpeg -> Binary Format File

## Type of Files
### 1. Text File -> .txt, .doc, etc
### 2. Binary File -> .png, .mp4, .mp3, .mov

```python
'''File Access Modes

 Mode    Name      Description                                  Behavior
 "r"     Read      Default mode. Opens a file for reading.      Throws FileNotFoundError if the file doesn't exist.
 "w"     Write     Opens a file for writing.                    Overwrites existing file content. Creates the file if missing.
 "x"     Exclusive Opens a file for exclusive creation.         Throws an error if the file already exists.
 "b"     Binary    Used for non-text files (images, PDFs).      Combined with others, e.g., "rb" or "wb".
'''
```

```python
# Read a file
# read() method reads the entire content of the file and returns it as a string.

f=open('example.txt', 'r')
content = f.read()
print(content)
f.close()
```

```python
# Write to a file
# write() method writes a string to the file. If the file already exists, it will overwrite the existing content.
f=open('example.txt', 'w')
f.write('This is an example of writing to a file. ')
f.close()
```

```python
# Append to a file
# append() method adds content to the end of the file without overwriting the existing content.
f=open('example.txt', 'a')
f.write('\nWe are learning about Gen AI and Python.')
f.close()
```

## With Statement - It automatically closes the file for you when the code block finishes, even if your program crashes or encounters an error

```python
with open('example.txt', 'r') as f:
    content = f.read()
    print(content)
```

```python
with open('example.txt', 'w') as f:
    f.write('We are exploring with statement in file handling in Python.')
```

```python
with open('example.txt', 'a') as f:
    f.write('\nThis is an example of using with statement for appending to a file.')
```

```python
lines = ['Line 1: This is the first line.\n', 'Line 2: This is the second line.\n', 'Line 3: This is the third line.\n']
with open('example.txt', 'w') as f:
    f.writelines(lines)
```

# Binary File Handling

```python
## Mode "wb" -> Write the content into the binary file
## Mode "rb" -> Read the content of the binary file
```

```python
# Write data to binary file
data=b'Hello, Binary World!'
with open("binary_data.txt","wb") as binary_data:
    binary_data.write(data)
```

```python
# Read data from binary file
with open("binary_data.txt","rb") as binary_data:
    content=binary_data.read()
    print("Content of Binary file")
    print(content)
```
