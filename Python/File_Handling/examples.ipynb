## readlines() 
### The readlines() function in Python reads all lines from a file and returns them as a list of strings

```python
# Count and print Total Lines in a file

with open('example.txt','r') as file:
    lines=file.readlines()
    total_lines=len(lines)
    print(f"Total Number of lines: {total_lines}")
    for line in lines:
        print(line, end="")

# print(lines)
```

```python
# Count and print Total words in a file

with open('example.txt','r') as file:
    lines=file.readlines()
    total_words=sum(len(line.split()) for line in lines)
    print(f"Total number of words in the file is : {total_words}\n") 
       
    for line in lines:
        line_words=line.split()
        print(line_words)
        print(f"Total number of words in the line is {len(line_words)}\n")
```

```python
# Count and print Total words in a file

with open('example.txt','r') as file:
    lines=file.readlines()
    total_words=sum(len(line) for line in lines)
    print(f"Total number of words in the file is {total_words}\n") 
```

## Seek Function
### seek() method is used to move the file cursor (pointer) to a specific position inside an open file

```python
with open('data_source.txt','w+') as file:
    file.write("Hello World!")
    file.seek(0)
    content1=file.read() 
    print(content1)
    file.seek(3)
    content2=file.read() 
    print(content2)   
       
    
```
