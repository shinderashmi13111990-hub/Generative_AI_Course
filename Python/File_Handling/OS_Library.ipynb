# OS Library

### OS Library is able to interact with our Operating System and can let us work with Multiple Directories and Files

```python
# getcwd() -> returns path of current working directory
import os
current_path=os.getcwd()
print("The current working directory is : ", current_path)
```

```python
# mkdir() -> creates directory 
new_path="New_Folder"
os.mkdir(new_path)
print(f"The new directory {new_path} is created successfullt")
```

```python
# listdir() -> Listing of all the directories present
all_items=os.listdir('.')
print(all_items)
```

```python
# Joining multiple paths together
dir_name="Folder_data"
file_name="File_data"
Full_path=os.path.join(dir_name,file_name)
print(f"The Full path is: {Full_path}")
```

```python
# Join Multiple path with the current path
current_complete_path=os.path.join(os.getcwd(),dir_name,file_name)
print(f"The complete path is : {current_complete_path}")
```

```python
# os.path.exists() -> check if the given path exists or not
given_path="Desktop\Rashmi\Generative AI Course\Python\app.py"
if os.path.exists(given_path):
    print("The path exists")
else:
    print("The path does not exists")
```

```python
# os.path.isfile() -> to check if the current path is a file or not
# os.path.isdir() -> to check if the current path is a directory or not

path="New_Folder"
if os.path.isfile(path):
    print("The current path is file")
elif os.path.isdir(path):
    print("The current path is directory")
else:
    print("The current path is invalid")
```

```python
# os.path.abspath() -> It provide us the complete path or the absolute path of the file or directory
file_name="binary_data.txt"
complete_path=os.path.abspath(file_name)
print("The complete path for the given file is ",complete_path)
```
